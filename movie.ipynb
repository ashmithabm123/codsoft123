```python
import pandas as pd
```

```python
movie = pd.read_csv('C:\\Users\\ASHMITHA\\Desktop\\codsoft\\IMDb.csv', encoding='latin1')
movie
```

```python
df = movie.dropna(subset=['Year','Duration','Genre','Rating','Votes','Director','Actor 1','Actor 2','Actor 3'])
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in ['Genre','Director','Actor 1','Actor 2','Actor 3']:
    df.loc[:, col] = LabelEncoder().fit_transform(df[col])
```

```python
df.loc[:, 'Duration'] = df['Duration'].astype(str).str.replace(' min', '').astype(float)
```

```python
df.loc[:, 'Votes'] = df['Votes'].astype(str).str.replace(',', '').astype(int)
```

```python
df.loc[:, 'Year'] = df['Year'].astype(str).str.extract(r'(\d{4})')[0].astype(int)
```

```python
x = df.drop(['Name','Rating'],axis=1)
x
```

```python
y = df.Rating
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python

model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_test,model.predict(x_test))
```
